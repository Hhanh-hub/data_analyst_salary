```python
## format salary:
#loc salary: thoa thuan
data.loc[data['Salary']=='Thoả thuận', 'Salary'] = np.nan
data[:5]
```

```python
data['Salary'] = data['Salary'].str.replace(' triệu','').str.replace('Tới ','').str.replace(' USD','').str.replace('Trên ','').str.replace(',','')
sl_2 = data['Salary'].str.split("-",expand=True)
```

```python
sl_2.head()
```

```python
data['Min Salary'] = sl_2[0]
data['Max Salary'] = sl_2[1]
data.drop(columns='Salary',inplace=True)
data.head()
```

```python
data.columns
```

```python
data_f = data.loc[:, ['Company', 'Title', 'Min Salary', 'Max Salary','Quantity', 'Working form', 'Level', 'Sex',
       'Experienced', 'Links']]
data_f.head()
```

```python
data_f.info()
```

```python
data_f[['Min Salary','Max Salary']] = data_f[['Min Salary','Max Salary']].fillna(0)
```

```python
data_f[['Min Salary','Max Salary']]= data_f[['Min Salary','Max Salary']].astype('float')
data_f.info()
```

```python
data_f['Min Salary'] = np.where(data_f['Min Salary']<500, round(data_f['Min Salary']*1000000,1), 
                                round(data_f['Min Salary']*23000,1))
```

```python
data_f['Max Salary'] = np.where(data_f['Max Salary']<500, round(data_f['Max Salary']*1000000,1), 
                                round(data_f['Max Salary']*23000,1))
```

```python
data_f['Quantity'] = data_f['Quantity'].str.extract('(\d)')
```

```python
data_f['Experienced'] = data_f['Experienced'].str.extract('(\d)')
```
